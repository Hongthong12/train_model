```python
import cv2 as cv
def image_processed(file_path):
    try:
        # Read the image file
        hand_img = cv.imread(file_path)
        if hand_img is None:
            raise ValueError("Failed to read the image file")

        # Image processing
        img_rgb = cv.cvtColor(hand_img, cv.COLOR_BGR2RGB)
        img_flip = cv.flip(img_rgb, 1)

        mp_hands = mp.solutions.hands
        hands = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.7)
        output = hands.process(img_flip)
        hands.close()

        if output.multi_hand_landmarks:
            data = output.multi_hand_landmarks[0]
            data = str(data)
            data = data.strip().split('\n')

            garbage = ['landmark {', '  visibility: 0.0', '  presence: 0.0', '}']
            without_garbage = [i for i in data if i not in garbage]

            clean = [i.strip()[2:] for i in without_garbage]
            clean = [float(i) for i in clean]

            return clean
        else:
            return np.zeros([1, 63], dtype=int)[0]

    except Exception as e:
        print(f"Error processing image: {e}")
        return None

def make_csv():
    mypath =  r"C:\Users\hfhon\OneDrive\Desktop\train_model\DATASET"
    file_name = open('dataset.csv', 'a')

    for each_folder in os.listdir(mypath):
        if '._' in each_folder:
            pass
        else:
            for each_number in os.listdir(mypath + '/' + each_folder):
                if '._' in each_number:
                    pass
                else:
                    label = each_folder
                    file_loc = os.path.join(mypath, each_folder, each_number)

                    # Check if the file exists and is an image
                    if os.path.isfile(file_loc) and file_loc.lower().endswith(('.png', '.jpg', '.jpeg')):
                        # Process the image
                        data = image_processed(file_loc)
                        if data is not None:
                            try:
                                for id, i in enumerate(data):
                                    if id == 0:
                                        print(i)
                                    file_name.write(str(i))
                                    file_name.write(',')
                                file_name.write(label)
                                file_name.write('\n')
                            except:
                                file_name.write('0')
                                file_name.write(',')
                                file_name.write('None')
                                file_name.write('\n')
                    else:
                        print(f"Invalid file: {file_loc}")

    file_name.close()
    print('Data Created !!!')

if __name__ == "__main__":
    make_csv()
```
